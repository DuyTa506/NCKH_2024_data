```python
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By

# Load your DataFrame
df = pd.read_csv('diendandulich.csv')
urls = df['dad-href'].tolist()

# Set up Chrome WebDriver
chrome_options = Options()
chrome_options.add_argument('--headless')  # Run Chrome in headless mode (without GUI)
driver = webdriver.Chrome(options=chrome_options)

with open('diendandulich.txt', 'a+', encoding='utf-8') as file:
    for url in urls:
        driver.get(url)
        
        driver.implicitly_wait(5)
        
        # Extract the content inside the specified div
        target_div = driver.find_element(By.CSS_SELECTOR, '#ctl00_MainContent_contentAr.content1')

        content_inside_div = target_div.text

        # Write the content to the file
        file.write(content_inside_div + '\n')
        file.write('=' * 50 + '\n')
driver.quit()
```

```python
import re
import pandas as pd
from urllib.parse import urljoin
from parsel import Selector
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import os

def clean_text(text):
    if text is not None:
        text = str(text)
        text = text.replace('\u200a', '')
        text = text.replace('\xa0', '')
        text = ' '.join(text.strip().split())
    return text

csv_file = "co_so_luu_tru.csv"

base_url = "https://csdl.vietnamtourism.gov.vn"
edge_options = webdriver.EdgeOptions()
edge_options.add_argument('--blink-settings=imagesEnabled=false')

edge_options.add_argument('--headless')
driver = webdriver.Edge(options=edge_options)
#chrome_options = Options()
# chrome_options.add_argument('--headless') 
# chrome_options.add_experimental_option(
#     # this will disable image loading
#     "prefs", {"profile.managed_default_content_settings.images": 2}
# )
# driver = webdriver.Chrome(options=chrome_options)
driver.get(base_url)
driver.implicitly_wait(2)

sel = Selector(text=driver.page_source)
full_urls = []

for item in sel.xpath("//div[contains(@class,'card')]/div[contains(@class, 'card-body')]/div[@class='embed-responsive embed-responsive-1by1 text-center']"):
    urls = item.css('a::attr(href)').get()
    urls = urls.split() if isinstance(urls, str) else urls
    full_urls.extend([url if url.startswith("http") else urljoin(base_url, url) for url in urls])

"""Crawl cơ sở lưu trú ."""
url_cslt = full_urls[0]
info_list = []
for page_number in range(99, 945):
    new_url = f"{url_cslt}?page={page_number}"
    driver.get(new_url) #co so luu tru
    driver.implicitly_wait(5)
    sel = Selector(text=driver.page_source)
    print("Crawling page :", page_number)
    for i, item in enumerate(sel.xpath("//div[contains(@class, 'verticle-listing-caption')]")):
    # Extract all information within the <div> element
        info = {
            'hotel_name': clean_text(item.xpath("h4/a/text()").get()),
            'type': clean_text(item.xpath("span[contains(@class, 'd-block')]/i[contains(@class, 'fa')]/following-sibling::text()").get()),
            'address': clean_text(item.xpath("span[contains(@class, 'd-block')]/i[contains(@class, 'fa fa-map-marker')]/following-sibling::text()").get()),
            'number_of_rooms': clean_text(item.xpath("span[contains(@class, 'd-block')]/i[contains(@class, 'fa fa-bed')]/following-sibling::text()").getall()[-1]),
            'phone': clean_text(item.xpath("span[contains(@class, 'd-block')]/i[contains(@class, 'fa fa-phone')]/following-sibling::text()").get()),
            'website': clean_text(item.xpath("span[contains(@class, 'd-block')]/i[contains(@class, 'fa fa-globe')]/following-sibling::text()").get()),
        }

        info_list.append(info)

    if os.path.exists(csv_file):
        existing_df = pd.read_csv(csv_file)
        df = pd.concat([existing_df, pd.DataFrame(info_list)], ignore_index=True)
    else:
        df = pd.DataFrame(info_list)

    
    df.to_csv(csv_file, index=False)
driver.quit()

```

```python
import re
import pandas as pd
from urllib.parse import urljoin
from parsel import Selector
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import os

def clean_text(text):
    if text is not None:
        text = str(text)
        text = text.replace('\u200a', '')
        text = text.replace('\xa0', '')
        text = ' '.join(text.strip().split())
    return text
edge_options = webdriver.EdgeOptions()
edge_options.add_argument('--blink-settings=imagesEnabled=false')

edge_options.add_argument('--headless')
driver = webdriver.Edge(options=edge_options)
with open("urls.txt", "r", encoding="utf-8") as url_file:
    base_urls = url_file.read().splitlines()

for i, base_url in enumerate(base_urls):
    driver.get(base_url)
    driver.implicitly_wait(3)
    item = Selector(text=driver.page_source)

    header = clean_text(item.xpath("//div[contains(@class, 'header')]/h4/a/text()").get())
    content_list = item.xpath("//div[contains(@class, 'col-12 py-2 content-detail')]/text()").getall()
    cleaned_content_list = [clean_text(content) for content in content_list]
    article = ' '.join(cleaned_content_list)
    all_p_content_list = item.xpath("//div[contains(@class, 'col-12 py-2 content-detail')]/p/text()").getall()
    all_content = ' '.join([clean_text(content) for content in all_p_content_list])

    print("Saving crawled" ,i+1)
    file_path = "diem_den_du_lich.txt"
    with open(file_path, "a", encoding="utf-8") as file:
        file.write(f"{header}\n\n")
        file.write(f"{article}\n\n")
        file.write(f"{all_content}\n{'=' * 50}\n")
```
