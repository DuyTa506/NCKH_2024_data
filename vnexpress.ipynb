```python
import re
import pandas as pd
from urllib.parse import urljoin
from parsel import Selector
import pandas as pd
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
import os

def clean_text(text):
    if text is not None:
        text = str(text)
        text = text.replace('\u200a', '')
        text = text.replace('\xa0', '')
        text = ' '.join(text.strip().split())
    return text

csv_file = "co_so_luu_tru.csv"

base_url = "https://csdl.vietnamtourism.gov.vn"
edge_options = webdriver.EdgeOptions()
edge_options.add_argument('--blink-settings=imagesEnabled=false')

edge_options.add_argument('--headless')
driver = webdriver.Edge(options=edge_options)
#chrome_options = Options()
# chrome_options.add_argument('--headless') 
# chrome_options.add_experimental_option(
#     # this will disable image loading
#     "prefs", {"profile.managed_default_content_settings.images": 2}
# )
# driver = webdriver.Chrome(options=chrome_options)
driver.get(base_url)
driver.implicitly_wait(2)

sel = Selector(text=driver.page_source)
full_urls = []

for item in sel.xpath("//div[contains(@class,'card')]/div[contains(@class, 'card-body')]/div[@class='embed-responsive embed-responsive-1by1 text-center']"):
    urls = item.css('a::attr(href)').get()
    urls = urls.split() if isinstance(urls, str) else urls
    full_urls.extend([url if url.startswith("http") else urljoin(base_url, url) for url in urls])

"""Crawl cơ sở lưu trú ."""
url_cslt = full_urls[0]
info_list = []
for page_number in range(99, 945):
    new_url = f"{url_cslt}?page={page_number}"
    driver.get(new_url) #co so luu tru
    driver.implicitly_wait(5)
    sel = Selector(text=driver.page_source)
    print("Crawling page :", page_number)
    for i, item in enumerate(sel.xpath("//div[contains(@class, 'verticle-listing-caption')]")):
    # Extract all information within the <div> element
        info = {
            'hotel_name': clean_text(item.xpath("h4/a/text()").get()),
            'type': clean_text(item.xpath("span[contains(@class, 'd-block')]/i[contains(@class, 'fa')]/following-sibling::text()").get()),
            'address': clean_text(item.xpath("span[contains(@class, 'd-block')]/i[contains(@class, 'fa fa-map-marker')]/following-sibling::text()").get()),
            'number_of_rooms': clean_text(item.xpath("span[contains(@class, 'd-block')]/i[contains(@class, 'fa fa-bed')]/following-sibling::text()").getall()[-1]),
            'phone': clean_text(item.xpath("span[contains(@class, 'd-block')]/i[contains(@class, 'fa fa-phone')]/following-sibling::text()").get()),
            'website': clean_text(item.xpath("span[contains(@class, 'd-block')]/i[contains(@class, 'fa fa-globe')]/following-sibling::text()").get()),
        }

        info_list.append(info)

    if os.path.exists(csv_file):
        existing_df = pd.read_csv(csv_file)
        df = pd.concat([existing_df, pd.DataFrame(info_list)], ignore_index=True)
    else:
        df = pd.DataFrame(info_list)

    
    df.to_csv(csv_file, index=False)
driver.quit()

```

```python
############### full url all_page ##########################

from bs4 import BeautifulSoup
from scrapy.selector import Selector
from selenium import webdriver

def crawl_page(url, filename):
    driver.get(url)
    driver.implicitly_wait(3)
    sel = Selector(text=driver.page_source)

    for item in sel.xpath("//div[contains(@class,'content-list-right content-list-center')]/div[contains(@class, 'header-list-style')]/h2[contains(@class, 'penci-entry-title entry-title grid-title')]") :
        urls = item.css('a::attr(href)').get()
        urls = urls.split() if isinstance(urls, str) else urls
        with open(filename, "a") as file:
            file.write(urls[-1] + "\n")

    html_content = sel.extract()
    soup = BeautifulSoup(html_content, 'html.parser')

    next_page_link = soup.find('a', class_='next')

    if next_page_link:
        next_page_url = next_page_link['href']
        print("URL trang tiếp theo:", next_page_url)
        return next_page_url
    else:
        print("Không có trang tiếp theo, chuyển qua URL kế tiếp.")
        return None

def crawl_for_base_urls(base_urls, output_filename):
    for base_url in base_urls:
        current_url = base_url
        while current_url:
            current_url = crawl_page(current_url, output_filename)



output_filename = "urls.txt"


edge_options = webdriver.EdgeOptions()
edge_options.add_argument('--blink-settings=imagesEnabled=false')
edge_options.add_argument('--headless')
driver = webdriver.Edge(options=edge_options)
crawl_for_base_urls(href_base, output_filename)

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from bs4 import BeautifulSoup
import time
from random import uniform

url = "https://www.klook.com/vi/blog/country/vietnam/?spm=Blog.DestinationLayer&clickId=07271d7a0d"

# Khởi tạo WebDriver
edge_options = webdriver.EdgeOptions()
edge_options.add_argument('--blink-settings=imagesEnabled=false')
# Uncomment the line below when running in production
# edge_options.add_argument('--headless')
driver = webdriver.Edge(options=edge_options)
driver.get(url)

wait = WebDriverWait(driver, 5)

all_hrefs = []

try:
    while True:
        # Lấy nội dung từ trang hiện tại
        html_content = driver.page_source
        soup = BeautifulSoup(html_content, 'html.parser')

        hrefs = [a['href'] for a in soup.find_all('a', href=True)]
        all_hrefs.extend(hrefs)

        next_button = driver.find_element(By.CLASS_NAME, "klk-pagination-next-btn")
        if next_button.is_enabled():
            next_button.click()

            wait.until(EC.staleness_of(driver.find_element(By.CLASS_NAME, "klk-pagination-item-current")))

            delay_time = uniform(3, 5)
            time.sleep(delay_time)
        else:
            break

except Exception as e:
    print(f"An exception occurred: {e}")

finally:

    driver.quit()

# Write URLs to a file
with open("urls.txt", "w", encoding="utf-8") as file:
    for href in all_hrefs:
        file.write(href + "\n")
```
